```python
import requests
import json

i = 0
iss_data = []
while(i<100):
    i += 1
    x = requests.get('http://api.open-notify.org/iss-now.json')
    y = x.text
    iss_data.append(y)

with open("./iss-data.json","w") as g:
    json.dump(iss_data, g)
```

```python
import matplotlib.pyplot as plt
import json

with open("./iss-data.json", "r") as json_file:
    x = json.loads(json_file.read())
    
latitude_list = []
longitude_list = []
    
for i,m in enumerate(x):
    latitude_list.append(m[53:60])
    longitude_list.append(m[78:85])
    
plt.scatter(latitude_list,longitude_list)
plt.title("International Space Station Location")
plt.xlabel("Latitude")
plt.ylabel("Longitude")

plt.show()
```
